```python
#import libraries for data analysis
import pandas as pd
import numpy as np

steam_df = pd.read_csv(r'C:\Users\Heisenberg\Downloads\games.csv')
#data exploration

steam_df.head(10)
```

```python
print(steam_df.info())
```

```python
#data cleaning to get month and year from Release date
steam_df['Release_Month']=steam_df['Release date'].str.split(' ').str[0]
steam_df['Release_Year']=steam_df['Release date'].str.split(' ').str[-1]
steam_df['Release_Year']=steam_df['Release_Year'].astype(int)
print(steam_df)
```

```python
#remove unnecessary columns to ensure clean and faster analysis in later stages 

removed_columns= ['Recommendations','Notes','About the game','DLC count','Screenshots','Movies','Website','Support email','Support url','Header image','Metacritic url','Average playtime two weeks','Median playtime forever','Median playtime two weeks','Reviews']

```

```python
steam_df = steam_df.drop(columns=removed_columns)

```

```python
steam_df.columns
```

```python
#taking only the upper bound for ease of analysis
def extract_upper_bound(range_str):
    #split the range and take the upper bound 
    try:
        upper_bound = range_str.split('-')[1].strip()
        return int(upper_bound)
    except IndexError: 
        return 0 
    
#applying the function to estimated owners

steam_df['upper_bound'] = steam_df['Estimated owners'].apply(extract_upper_bound)

```

```python
print(steam_df)
```

```python
# we can see from above that there are many games with 0 as upper bound value therefore we will verify the count for those games
steam_df['Estimated owners'].value_counts()

```

```python
# seeing that many upper bounds are 0 as well, we will assume that many games were only test games and filter them

steam_df = steam_df[steam_df['Estimated owners'] != '0 - 0']
```
